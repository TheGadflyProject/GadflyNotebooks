```python
import numpy as np
import pandas as pd
```

```python
shakes_url = "http://courses.ischool.berkeley.edu/i290-dds/s16/data/shakes.txt"
shakes = pd.read_table(shakes_url)
```

```python
shakes
```

```python
ac_mean = shakes["AbstractConcepts"].mean()
fp_mean = shakes["FirstPerson"].mean()
print("AC Mean: {}      FP Mean: {}".format(ac_mean, fp_mean))
```

```python
x = shakes["AbstractConcepts"] - ac_mean
y = shakes["FirstPerson"] - fp_mean
print(sum(x*y) / (len(shakes) -1))
```
